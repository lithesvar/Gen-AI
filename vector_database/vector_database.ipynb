```python
import chromadb
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma

docs = ["Apple", "Banana", "Friendship", "Love", "Hate", "Australia", "Japan"]
client = chromadb.PersistentClient(path=".") #create vectorDB in current dir
db = Chroma.from_texts(docs, OpenAIEmbeddings(), client=client)
```

```python
db.similarity_search_with_score('Emotion', k=3)
```

```python
db.similarity_search_with_score('country', k=3)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
docs = ["Apple", "Banana", "Friendship", "Love", "Hate", "Australia", "Japan","Eye","Nose","Heart", "India", "Taj Mahal"]
```

```python
embeddings = OpenAIEmbeddings()
db = FAISS.from_texts(docs, embeddings)
print(db.index.ntotal)
```

```python
db.similarity_search("body",3)
```
